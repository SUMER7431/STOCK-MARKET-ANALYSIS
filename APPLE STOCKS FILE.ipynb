```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
apple = yf.download('AAPL', start='2020-01-01', end='2024-01-01')
print(apple.shape)
print(apple.head())
```

```python
apple['Close'].plot(figsize=(10,5), title='Apple Stock Price 2020-2024')
plt.ylabel('Price USD')
plt.show()
```

```python

apple['MA50'] = apple['Close'].rolling(50).mean()
apple['MA200'] = apple['Close'].rolling(200).mean()
apple[['Close','MA50','MA200']].plot(figsize=(10,5), title='Apple Moving Averages')
plt.show()
```

```python
stocks = yf.download(['AAPL','GOOGL','MSFT','AMZN'], start='2020-01-01', end='2024-01-01')
stocks['Close'].plot(figsize=(10,5), title='Tech Stocks Comparison')
plt.show()
```

```python
apple['Daily Return'] = apple['Close'].pct_change()
apple['Daily Return'].plot(kind='hist', bins=50, title='Apple Daily Returns Distribution')
plt.show()
```

```python
apple['Year'] = apple.index.year
yearly = apple.groupby('Year')['Close'].mean()
yearly.plot(kind='bar', title='Apple Average Price by Year', color='steelblue')
plt.show()
```
